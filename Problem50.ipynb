```
def check_prime(n):
  flag=1
  if n==1:
    flag=0
  if n>1:
    for i in range(2,n):
      if n%i==0:
        flag=0
        break
  return flag
  
```

```
start=1
end=1000
```

```
def list_prime(s,e):
  prime_ls=[]
  for i in range(s,e):
    res=check_prime(i)
    if res==1:
      prime_ls.append(i)
  return prime_ls
    
    
```

```
ls_prime=list_prime(start,end)
ls_prime
```

```

sum_list_list=[]
count_ls_list=[]


for i in range(0,len(ls_prime)):
  total=0
  sum_list=[]
  count_ls=[]
  count=0
  for j in range(i,len(ls_prime)):
    
    total=total+ls_prime[j]
    ck=check_prime(total)
    count=count+1
    if ck==1 and total<end:
      sum_list.append(total)
      count_ls.append(count)
  sum_list_list.append(sum_list)
  count_ls_list.append(count_ls)
print(sum_list_list)
print(count_ls_list)


    
    
```

```
flat_list1 = []
flat_list2= []
for sublist in sum_list_list:
    for item in sublist:
        flat_list1.append(item)
print(flat_list1)
for sublist in count_ls_list:
    for item in sublist:
        flat_list2.append(item)
print(flat_list2)
```
