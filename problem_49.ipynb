```
def check_prime(n):
  flag=1
  if n>999:
    for i in range(2,int(n/2)):
      if n%i==0:
        flag=0
        break
  return flag
```

```
prime_numbers=[]
for i in range (1000,10001):
  res=check_prime(i)
  if res==1:
    prime_numbers.append(i)
print(prime_numbers)
```

```

num_list=[]
for l in prime_numbers:
  dig_list=[]
  q1=l/1000
  dig_list.append(int(q1))
  r1=l%1000
  q2=r1/100
  dig_list.append(int(q2))
  r2=r1%100
  q3=r2/10
  dig_list.append(int(q3))
  r3=r2%10
  dig_list.append(int(r3))
  num_list.append(dig_list)
num_list





```

```
import itertools
fl=[]
for n in num_list:
  li2=list(itertools.permutations(n,4))
  fl.append(li2)
fl
  
    
```
